```python
# Import library
import pandas as pd
```

```python
# Import dataset
titanic = "https://raw.githubusercontent.com/datasciencedojo/datasets/refs/heads/master/titanic.csv"
df = pd.read_csv(titanic)
df.head()
```

```python
# Create table contain unique value from survived, pclass, sex, cabin, and embarked field 
categorical_unique_value = {
    'survived': [df['Survived'].unique()],
    'pclass': [df['Pclass'].unique()],
    'sex': [df['Sex'].unique()],
    'cabin': [df['Cabin'].unique()],
    'embarked': [df['Embarked'].unique()],
}

pd.DataFrame(categorical_unique_value).T\
    .rename({0 : 'unique_values'}, axis = 1)
```

### Column Details
- `PassengerId`: Passanger Id
- `Survived`: Passanger Survived Status (0 = Not Survived ; 1 = Survived)
- `Pclass`: Passenger Class (1 = Upper/VIP ; 2 = Middle ; 3 = Lower)
- `Name`: Passenger Name
- `Sex`: Passenger Gender (male ; female)
- `Age`: Passenger Age
- `SibSp`: Number of Siblings/Spouses Aboard in The Titanic
- `Parch`: Number of Parents/Children Aboard in The Titanic
- `Ticket`: Ticket Id 
- `Fare`: Ticket Price
- `Cabin`: Passenger Cabin Number 
- `Embarked`: Port of Departure (S = Southampton, C = Cherbourg, Q = Queenstown)

# Task 2 — Querying dengan kondisi majemuk

```python
# Get passenger count with following condition:
# - Cabin class 1 or 2
# - Age > 30
# - Not Survived (Survived = 0)

df[
    (df['Pclass'].isin([1, 2])) \
      & (df['Age'] > 30) \
      & (df['Survived'] == 0)
]['PassengerId'].count()
```

```python
# Get passengers with following condition:
# - Female
# - Survived (Survived = 1)
# - Age < 18

df[
    (df['Sex'] == 'female') \
    & (df['Survived'] == 1)\
    & (df['Age'] < 18)
][['Name', 'Age', 'Pclass', 'Embarked']]
```

```python
# Define age classification
# Age <= 12 = "Anak"
# Age 13 - 17 = "Remaja"
# Age 18 - 59 = "Dewasa"
# Age >= 60 = "Lansia" 

df['kelompok_usia'] = pd.cut(df['Age'],
                             bins=[0, 12, 17, 59, max(df['Age'])], 
                             labels=['Anak', 'Remaja', 'Dewasa', 'Lansia'])
df
```

# Task 3 — Groupby dan agregasi

```python
# Calculate survival rate 
# Survival rate = Count of Survived / Total of Passenger

# Get count of survived passenger group by pclass
survived_passenger = df[df['Survived'] == 1][['Pclass','Survived']]\
                        .groupby('Pclass')\
                        .count()

survived_passenger['survival_rate (%)'] = round(survived_passenger['Survived'] / df['Survived'].count(), 2)
survived_passenger
```

```python
# Get Age and Fare average group by pclass and sex
df[['Pclass', 'Sex', 'Fare', 'Age']]\
    .groupby(['Pclass', 'Sex'])\
    .agg({'Age':'mean', 'Fare':'mean'})\
    .round({'Age': 0})
```

```python
# Get count of pessanger survived / not survived group by kelompok_usia
pd.crosstab(df['kelompok_usia'], df['Survived'])\
    .rename({0:'Not Survived', 1:'Survived'}, axis = 1)
```
