```python
import pandas as pd
```

```python
titanic = "https://raw.githubusercontent.com/datasciencedojo/datasets/refs/heads/master/titanic.csv"
df = pd.read_csv(titanic)
df.head()
```

```python
df.groupby(['Sex','Pclass']).agg({'Fare':['min','median','max']})
```

```python
pd.crosstab(df['Survived'],df['Pclass'],df['Fare'],aggfunc='mean')
```
