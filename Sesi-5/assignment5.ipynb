### Global Variable

```python
# ======================================Task 1================================
menu:dict = {
    "Espresso": 25000,
    "Latte": 35000,
    "Cappuccino": 38000,
    "Croissant": 28000,
    "Sandwich": 42000,
    "Cheesecake": 45000,
}

transaksi:list[dict] = [
    {"nama": "Budi", "pesanan": ["Latte", "Croissant", "Croissant"]},
    {"nama": "Sari", "pesanan": ["Espresso", "Cheesecake"]},
    {"nama": "Dian", "pesanan": ["Cappuccino", "Sandwich", "Latte", "Cheesecake"]},
    {"nama": "Reza", "pesanan": ["Espresso"]},
    {"nama": "Ayu",  "pesanan": ["Latte", "Latte", "Sandwich", "Croissant", "Cheesecake"]},
]

# ======================================Task 2================================
pendapatan_cafe:int = 0
tagihan_pelanggan:dict = {}

# ======================================Task 3================================
antrian:list[str] = ["Budi", "Sari", "Dian", "Reza", "Ayu"]
stok_kopi:int = 3
```

### Define Function

```python
# ======================================Task 1================================
def count_discount(harga:int) -> dict:
    """
    Memberikan Detail Diskon.
    Arg:
        harga (int): Harga barang sebelum diskon.
    Return:
        dict: Detail diskon, seperti
              total_akhir, harga_diskon, presentase_diskon.
    """
    if harga < 50_000: # Jika harga < 50K, maka diskon 0%
        diskon = 0
        return { 
                "total_akhir": harga * (100 / 100), 
                "harga_diskon": harga - (harga * (100 / 100)), 
                "presentase_diskon": diskon * 100
            }
    elif (harga >= 50_000) and (harga <= 99_999): # Jika harga >= 50K dan <= 99K, maka diskon 10%
        diskon = 0.10
        return { 
                "total_akhir": harga * (90 / 100), 
                "harga_diskon": harga - (harga * (90 / 100)), 
                "presentase_diskon": diskon * 100
            }
    else: # Jika harga >= 100K, maka diskon 15%
        diskon = 0.15
        return { 
                "total_akhir": harga * (85 / 100), 
                "harga_diskon": harga - (harga * (85 / 100)), 
                "presentase_diskon": diskon * 100
            }


def get_price(list_pesanan:list[str]) -> int:
    """
    Kalkulasi harga yang harus pelanggan bayar.
    Arg:
        list_pesanan (int): List pesanan pelanggan.
    Return:
        int: Total harga yang harus pelanggan bayar. 
    """
    total:int = 0
    for m in list_pesanan:
        total += menu[str(m)]
    return total

# ======================================Task 2================================
def update_tagihan_pelanggan(nama_pelanggan:str, tagihan:int) -> None:
    """
    Update nominal tagihan tiap pelanggan.
    Args:
        nama_pelanggan (str): Nama pelanggan.
        tagihan (int): Total tagihan tiap pelanggan.
    """
    tagihan_pelanggan[str(nama_pelanggan)] = tagihan

def update_pendapatan_cafe(tagihan:int) -> None:
    """
    Update pendapatan kafe.
    Arg:
        tagihan (int): Total tagihan seluruh pelanggan.
    """
    pendapatan_cafe += tagihan

def get_tagihan_tertinggi() -> tuple[str, int]:
    """
    Mencari pelanggan dengan tagihan tertinggi.
    Return:
        tuple[str, int]: Pelanggan dengan tagihan tertinggi. 
                         Nama, Tagihan. 
    """
    thres_tagihan:int = 0
    thres_nama:str = ''
    for nama, tagihan in tagihan_pelanggan.items(): # Mencari tagihan maksimum
        if tagihan >= thres_tagihan:
            thres_tagihan = tagihan
            thres_nama = nama
        else:
            continue
    
    return thres_nama, thres_tagihan

# ======================================Task 3================================

def find_transaksi(nama_pelanggan:str) -> dict:
    """
    Mencari nama pelanggan di data transaksi.
    Arg:
        nama_pelanggan (str): Nama pelanggan yang ingin dicari.
    Return:
        dict: Data transaksi dari pelanggan yang dicari. 
    """
    for t in transaksi:
        if nama_pelanggan == t['nama']:
            return t
```

# Task 1 — Hitung tagihan per pelanggan

```python
for t in transaksi:
    harga_total:int = get_price(t['pesanan'])
    calculation:dict = count_discount(harga_total)
    update_tagihan_pelanggan(t['nama'], int(calculation['total_akhir']))
    update_pendapatan_cafe(calculation['total_akhir'])
    
    print(f"Pelanggan: {t['nama']}")
    print(f"Pesanan: {", ".join(t['pesanan'])}")
    print(f"Diskon: {int(calculation['presentase_diskon'])}% (Rp {int(calculation['harga_diskon']):,})")
    print(f"Subtotal: Rp {int(harga_total):,}")
    print(f"Total: Rp {int(calculation['total_akhir']):,}")
    print("="*100)

print("="*100)
print(f"Pendapatan kafe: Rp {int(pendapatan_cafe):,}")
```

# Task 2 — Cari pelanggan dengan tagihan tertinggi

```python
nama, tagihan = get_tagihan_tertinggi()
print("="*5 + "PELANGGAN TAGIHAN TERTINGGI" + "="*5)
print(f"Nama: {nama}")
print(f"Tagihan: Rp {tagihan:,}")
```

# Task 3 — Simulasi antrian dengan while loop

```python
n:int = 0
while n <= len(antrian):
    nama_pelanggan:str = str(antrian[0])
    pelanggan:dict = find_transaksi(nama_pelanggan)
    
    # Berhenti jika stok kopi == 0 (habis)
    if stok_kopi == 0:
        print("Stok kopi habis")
        break   
    
    # Kurangi stok kopi jika pelanggan pesan (Espresso, Latte, Cappucino)
    if ('Espresso' in pelanggan['pesanan']) \
        or ('Latte' in pelanggan['pesanan']) \
        or ('Cappuccino' in pelanggan['pesanan']):
        stok_kopi -= 1
    
    print(f"Melayani: {nama_pelanggan} | Stok kopi tersisa: {stok_kopi}")
    antrian.pop(0)
    
    n += 1

else:
    print("Semua pelanggan telah dilayani")
    
```
