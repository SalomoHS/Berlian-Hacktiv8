```python
# Import Library
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load dataset
raw_df = pd.read_excel('./Online Retail.xlsx')
df = raw_df.copy()
raw_df.head()
```

### Max

```python
# Print max for quantity and unit price field
print(f"Max Quantity: {df['Quantity'].max()}")
print(f"Max UnitPrice: {df['UnitPrice'].max()}")
```

### Min

```python
# Print min for quantity and unit price field
print(f"Min Quantity: {df['Quantity'].min()}")
print(f"Min UnitPrice: {df['UnitPrice'].min()}")
```

### Range

```python
# Print range for quantity and unit price field
print(f"Range Quantity: {df['Quantity'].max() - df['Quantity'].min()}")
print(f"Range UnitPrice: {df['UnitPrice'].max() - df['UnitPrice'].min()}")
```

### Std Dev

```python
# Print standard deviasion for quantity and unit price field
print(f"StdDev Quantity: {np.std(df['Quantity'])}")
print(f"StdDev UnitPrice: {np.std(df['UnitPrice'])}")
```

### Variance

```python
# Print variance for quantity and unit price field
print(f"Variance Quantity: {np.var(df['Quantity'])}")
print(f"Variance UnitPrice: {np.var(df['UnitPrice'])}")
```

### IQR

```python
# Print IQR for quantity and unit price field
unitprice_q3 = df['UnitPrice'].quantile(0.75)
unitprice_q1 = df['UnitPrice'].quantile(0.25)
print(f"UnitPrice IQR: {unitprice_q3 - unitprice_q1}")
quantity_q3 = df['Quantity'].quantile(0.75)
quantity_q1 = df['Quantity'].quantile(0.25)
print(f"Quantity IQR: {quantity_q3 - quantity_q1}")
```

### Percentile

```python
# Print percentile(10,25,75,90) for quantity and unit price field
print(f"UnitPrice Percentile ke-10: {df['UnitPrice'].quantile(0.1)}")
print(f"UnitPrice Percentile ke-25: {df['UnitPrice'].quantile(0.25)}")
print(f"UnitPrice Percentile ke-75: {df['UnitPrice'].quantile(0.75)}")
print(f"UnitPrice Percentile ke-90: {df['UnitPrice'].quantile(0.9)}")
print()
print(f"Quantity Percentile ke-10: {df['Quantity'].quantile(0.1)}")
print(f"Quantity Percentile ke-25: {df['Quantity'].quantile(0.25)}")
print(f"Quantity Percentile ke-75: {df['Quantity'].quantile(0.75)}")
print(f"Quantity Percentile ke-90: {df['Quantity'].quantile(0.9)}")
```

### Boxplot

```python
# Create boxplot for quantity and unit price field
fix, ax = plt.subplots(1,2,figsize=(12,4))
sns.boxplot(df,y='UnitPrice',ax=ax[0])
sns.boxplot(df,y='Quantity',ax=ax[1])

ax[0].set_ylabel('')
ax[1].set_ylabel('')

ax[0].set_title('Unit Price')
ax[1].set_title('Quantity')
```

### Violin

```python
# Create violin plot for quantity and unit price field
fix, ax = plt.subplots(1,2,figsize=(12,4))
sns.violinplot(df,y='UnitPrice',ax=ax[0])
sns.violinplot(df,y='Quantity',ax=ax[1])

ax[0].set_ylabel('')
ax[1].set_ylabel('')

ax[0].set_title('Unit Price')
ax[1].set_title('Quantity')
```
