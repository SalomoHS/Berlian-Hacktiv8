# Task 1 — Class Karyawan (base class)

```python
class Karyawan:
    def __init__(self, nama:str, department:str, gaji_pokok:int) -> None:
        self.nama:str = nama
        self.department:str = department
        self.__gaji_pokok:int = gaji_pokok

    def get_gaji_pokok(self) -> int:
        return self.__gaji_pokok

    def set_gaji_pokok(self, new_gaji_pokok:int) -> str:
        if new_gaji_pokok > 0:
            self.__gaji_pokok = new_gaji_pokok
            return "SUCCESS: Gaji pokok berhasil di perbarui"
        else:
            return f"ERROR: Gaji pokok harus lebih dari 0"

    def hitung_gaji(self):
        self.__total_gaji:int = self.__gaji_pokok
        return f"Gaji pokok: {self.__total_gaji}"

    def info(self):
        return {
            "Nama": self.nama,
            "Department": self.department,
            "Gaji Pokok": self.__gaji_pokok,
            "Total Gaji": self.hitung_gaji()
        }
```

```python
firman:Karyawan = Karyawan('Firman', 'Marketing', 10000000)
firman.info()
```

# Task 2 — Subclass KaryawanTetap

```python
class KaryawanTetap(Karyawan):
    def __init__(self, nama:str, department:str, gaji_pokok:int, 
                 tunjangan_tetap:int, bonus_tahunan):int:

        super().__init__(nama, department, gaji_pokok)
        self.tunjangan_tetap:int = tunjangan_tetap
        self.bonus_tahunan:int = bonus_tahunan
        

    def hitung_gaji(self) -> int:
        self.__total_gaji:int = int( 
                self.get_gaji_pokok() \
                + self.tunjangan_tetap \
                + (self.bonus_tahunan / 12)
            )
        return self.__total_gaji
    
    def info(self):
        return super().info() | {
            "Jenis Karyawan": "Karyawan Tetap",
            "Tunjangan": self.tunjangan_tetap,
            "Bonus /bulan": int(self.bonus_tahunan / 12)
        }
```

```python
cahya:KaryawanTetap = KaryawanTetap("cahya", "IT", 12000000, 2000000, 8000000)
rai:KaryawanTetap = KaryawanTetap("rai", "IT", 13000000, 2000000, 8000000)

print(("=" * 10) + " Karyawan 1 " + ("=" * 10)) 
print(f"Jenis Karyawan: {cahya.info()['Jenis Karyawan']}")
print(f"Tunjangan: Rp {cahya.info()['Tunjangan']:,}")
print(f"Bonus (per bulan): Rp {cahya.info()['Bonus /bulan']:,}")

print()

print(("=" * 10) + " Karyawan 2 " + ("=" * 10)) 
print(f"Jenis Karyawan: {rai.info()['Jenis Karyawan']}")
print(f"Tunjangan: Rp {rai.info()['Tunjangan']:,}")
print(f"Bonus (per bulan): Rp {rai.info()['Bonus /bulan']:,}")

```

# Task 3 — Subclass KaryawanKontrak

```python
class KaryawanKontrak(Karyawan):
    def __init__(self, nama:str, department:str, durasi_kontrak_bulan:int, 
                 rate_per_jam:int, jam_per_bulan:int):
        super().__init__(nama, department, gaji_pokok = 0)
        self.durasi_kontrak_bulan:int = durasi_kontrak_bulan
        self.rate_per_jam:int = rate_per_jam
        self.jam_per_bulan:int = jam_per_bulan
        
    
    def hitung_gaji(self) -> int:
        self.__total_gaji:int = int(self.rate_per_jam * self.jam_per_bulan)
        return self.__total_gaji
    
    def info(self) -> dict:
        return super().info() | {
            "Jenis Karyawan": "Karyawan Kontrak",
            "Durasi Kontrak": self.durasi_kontrak_bulan,
            "Rate (per jam)": self.rate_per_jam,
            "Jam (per bulan)": self.jam_per_bulan
        }
```

```python
gina:KaryawanKontrak = KaryawanKontrak("gina", "HR", 12, 200000, 40)
rani:KaryawanKontrak = KaryawanKontrak("rani", "HR", 12, 200000, 40)

print(("=" * 10) + " Karyawan 1 " + ("=" * 10)) 
print(f"Jenis Karyawan: {gina.info()['Jenis Karyawan']}")
print(f"Durasi Kontrak: {gina.info()['Durasi Kontrak']:,} bulan")
print(f"Rate (per jam): Rp {gina.info()['Rate (per jam)']:,}")
print(f"Jam (per bulan): {gina.info()['Jam (per bulan)']}")

print()

print(("=" * 10) + " Karyawan 2 " + ("=" * 10))
print(f"Jenis Karyawan: {rani.info()['Jenis Karyawan']}")
print(f"Durasi Kontrak: {rani.info()['Durasi Kontrak']:,} bulan")
print(f"Rate (per jam): Rp {rani.info()['Rate (per jam)']:,}")
print(f"Jam (per bulan): {rani.info()['Jam (per bulan)']}")
```
