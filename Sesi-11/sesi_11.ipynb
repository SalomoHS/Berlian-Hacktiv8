# Filter Quantity and Unit Price > 0

```python
# Take only where quantity > 0 and unit price > 0
df = df[df['Quantity'] > 0]
df = df[df['UnitPrice'] > 0]
```

```python
# Plotting null values after filtering
plt.figure(figsize=(10,4))
sns.barplot(df.isnull().sum())
plt.title("Null Values")
```

```python
# Drop null value based on customer id field
df_no_null = df.dropna(subset="CustomerID")
df_no_null.isna().sum()
```

```python
plt.scatter(df_no_null['Quantity'],df_no_null['UnitPrice'])
plt.xlabel('Quantity')
plt.ylabel('UnitPrice')
```

```python
# Get data where unit price > 1000
df_no_null[df_no_null['UnitPrice'] > 1000]
```

```python
# Remove data where stock code is ambigous
df_exclude_non_product = df_no_null[~df_no_null['StockCode'].astype(str).str.match(r'^[a-zA-Z]+$')]
df_exclude_non_product
```

# Recency

```python
customer_recency = df.groupby('CustomerID',as_index=False)["InvoiceDate"].max()
recent_date = df["InvoiceDate"].max()
customer_recency['Recency'] = customer_recency.apply(lambda x: (recent_date - x['InvoiceDate']).days, axis=1)
customer_recency
```

# Monetary

```python
# Bikin kolom total price: quantity * unit price
df_exclude_non_product['TotalPrice'] = df_exclude_non_product['Quantity'] * df_no_null['UnitPrice']
```

```python
# Get customer monetary, then normalize the value (0 - 100)
customer_monetary = df_exclude_non_product.groupby('CustomerID',as_index=False)["TotalPrice"].sum()
customer_monetary['M_norm'] = (customer_monetary['TotalPrice'].rank() / customer_monetary['TotalPrice'].rank().max()) *100
customer_monetary
```

```python
# get 20% (assume as top spender)
thres = customer_monetary['M_norm'].quantile(0.8)
customer_monetary_nozero_20 = customer_monetary[customer_monetary['M_norm'] > thres]
customer_monetary_nozero_20
```

```python
# Filter data only show top spender
df_exclude_non_product = df_exclude_non_product.set_index('CustomerID')
customer_20_monetary = df_exclude_non_product.loc[customer_monetary_nozero_20['CustomerID']]
```

```python
# Get Product count order, total spend, and total quantity sold
product_count = customer_20_monetary.reset_index().groupby('Description', as_index=False).agg({'CustomerID':'nunique', 'TotalPrice':'sum', 'Quantity':'sum'}).sort_values('CustomerID', ascending=False)
product_count
```

```python
# Plot top 10 product bough by top spender
top_10_buy = product_count.sort_values('CustomerID',ascending=False).iloc[:10,] 
sns.barplot(top_10_buy, x='CustomerID', y= 'Description', orient='h')
plt.xlabel('Order Count')
```

```python
# Plot Top 10 Products by Customer Count & Total Price 
top_10_buy = product_count.sort_values('TotalPrice', ascending=False).iloc[:10]

fig, ax1 = plt.subplots(figsize=(10, 7))

y = np.arange(len(top_10_buy))
height = 0.4

# --- Bottom axis: CustomerID ---
bars1 = ax1.barh(y - height/2, top_10_buy['CustomerID'], height=height,
                 color='steelblue', label='CustomerID')
ax1.set_xlabel('CustomerID (Count)', color='steelblue')
ax1.tick_params(axis='x', labelcolor='steelblue')
ax1.set_yticks(y)
ax1.set_yticklabels(top_10_buy['Description'])
ax1.invert_yaxis()  # highest TotalPrice at the top

# --- Top axis: TotalPrice ---
ax2 = ax1.twiny()
bars2 = ax2.barh(y + height/2, top_10_buy['TotalPrice'], height=height,
                 color='salmon', alpha=0.8, label='TotalPrice')
ax2.set_xlabel('Total Price', color='salmon')
ax2.tick_params(axis='x', labelcolor='salmon')

# --- Legend ---
ax1.legend(handles=[bars1, bars2],
           labels=['CustomerID', 'TotalPrice'],
           loc='lower right')

plt.title('Top 10 Products by Customer Count & Total Price')
plt.tight_layout()
plt.show()
```
