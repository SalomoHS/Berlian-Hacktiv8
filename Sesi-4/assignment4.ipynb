```python
produk:list = [
    {"nama": "Laptop ASUS A516", "harga": 7500000, "stok": 12, "kategori": "laptop"},
    {"nama": "Mouse Logitech M100", "harga": 85000, "stok": 50, "kategori": "aksesoris"},
    {"nama": "Keyboard Mechanical Rexus", "harga": 450000, "stok": 30, "kategori": "aksesoris"},
    {"nama": "Monitor LG 24 inch", "harga": 2300000, "stok": 8, "kategori": "monitor"},
    {"nama": "SSD Kingston 512GB", "harga": 650000, "stok": 25, "kategori": "storage"},
]
```

# Task 1 — Informasi produk terformat

```python
for prod in produk:
    print(f"[{prod['kategori'].upper()}] {prod['nama']} | Harga: Rp {prod['harga']:,} | Stok: {prod['stok']} unit".replace(",", "."))
```

# Task 2 — Kalkulasi nilai inventory

```python
total_nilai_inventory:int = 0
total_stok:int = 0
for prod in produk:
    total_nilai_inventory += (prod['harga'] * prod['stok'])
    total_stok += prod['stok']

rata_harga:float = total_nilai_inventory/total_stok

print(f"Total Nilai Inventory: Rp {total_nilai_inventory:,}".replace(",","."))
print(f"Rata rata harga: {rata_harga:.2f}")
```

# Task 4 — Ringkasan per kategori

```python
ringkasan_kategori:dict = {}

for prod in produk:
    kategori = prod['kategori']
    ringkasan_kategori[str(kategori)] = 0

for prod in produk:
    kategori = prod['kategori']
    ringkasan_kategori[str(kategori)] += 1

for a,b in ringkasan_kategori.items():
    print(f"Kategori [{a.upper()}]: {b} produk")
```

```python
class Produk:
    def add_product(self, new_produk:dict) -> None:
        self.add_kategori(new_produk)
        produk.append(new_produk)
    
    def add_kategori(self, new_produk) -> None:
        if new_produk['kategori'] not in ringkasan_kategori.keys():
            ringkasan_kategori[str(new_produk['kategori'])] = 1
        else:
            ringkasan_kategori[str(new_produk['kategori'])] += 1
    
    def print_data(self) -> None:

        # Print Updated Produk List
        print("="*100)
        for idx, i in enumerate(produk):
            if idx == len(produk)-1:
                print("New Product -> " + str(i))
            else:
                print(" " * 15 + str(i))

        # Print Updated Ringkasan Kategori
        print("")
        print("="*100)
        for a,b in ringkasan_kategori.items():
            if a == new_produk['kategori']:
                print(f"{"Updated ->"} Kategori [{a.upper()}]: {b} produk")
            else:
                print(f"{" "*10} Kategori [{a.upper()}]: {b} produk")



new_produk:dict = {
    "nama": "Webcam Logitech C920", 
    "harga": 750000, 
    "stok": 15, 
    "kategori": "aksesoris"
}

_produk:Produk = Produk()
_produk.add_product(new_produk)
_produk.print_data()
```
